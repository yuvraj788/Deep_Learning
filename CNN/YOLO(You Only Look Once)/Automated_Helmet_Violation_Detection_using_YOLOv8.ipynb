```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")  # nano version (fastest)

model.train(
    data="data.yaml",
    epochs=50,
    imgsz=640,
    batch=16
)
```

```python
from ultralytics import YOLO

model = YOLO("/content/runs/detect/train/weights/best.pt")

results = model.predict(
    source="/root/.cache/kagglehub/datasets/chinmay2830/yolov7-helmet-detection-dataset/versions/1/helmet_dataset/test/images",
    conf=0.5,
    save=True
)
```
